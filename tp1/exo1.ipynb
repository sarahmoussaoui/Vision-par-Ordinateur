```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Charger une image en niveau de gris
img = cv2.imread('imgGris.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Vérifier si le chargement s'est fait correctement
if img is None:
    print("Erreur: L'image n'a pas pu être chargée.")
else:
    # 3. Créer une nouvelle image du même taille que img
    imgResultat = np.zeros_like(img)

    # 4. Inverser les valeurs des pixels (255 - niveau de gris de img)
    imgResultat = 255 - img

    # 5. Afficher img et imgResultat
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Image Originale (Gris)")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(imgResultat, cmap='gray')
    plt.title("Image Inversée (Gris)")
    plt.axis('off')

    plt.show()



# Même chose avec une autre photo
# 1.Charger une image couleur
img_color = cv2.imread('cat_large.jpg', cv2.IMREAD_COLOR)

# 2. Vwrifier si le chargement s'est fait correctement
if img_color is None:
    print("Erreur: L'image couleur n'a pas pu être chargée.")
else:
    # 3. Créer une nouvelle image du même taille que img_color
    #imgResultat_color = np.zeros_like(img_color)

    # 4. Inverser les valeurs des pixels pour chaque canal de couleur
    imgResultat_color = 255 - img_color

    # 5. Afficher img_color et imgResultat_color
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
    plt.title("Image Originale (Couleur)")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(imgResultat_color, cv2.COLOR_BGR2RGB))
    plt.title("Image Inversée (Couleur)")
    plt.axis('off')

    plt.show()
```

```python
# chaging type of image to uint32 and displaying it
img_uint32 = img.astype(np.uint32)
print(f"Type de l'image originale: {img.dtype}, Type de l'image convertie: {img_uint32.dtype}")
plt.imshow(img_uint32, cmap='gray')
```

```python
# displaying image in float64
img_float64 = img.astype(np.float64) / 255.0
print(f"Type de l'image convertie en float64: {img_float64.dtype}")
plt.imshow(img_float64, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Charger une image en niveau de gris
img = cv2.imread('images/imgGris.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Vrifier si le chargement s'est fait correctement
if img is None:
    print("Erreur: L'image n'a pas pu tre charge.")
else:
    # 3. Crer une nouvelle image du mme taille que img
    imgResultat = np.zeros_like(img)

    # 4. Inverser les valeurs des pixels (255 - niveau de gris de img)
    imgResultat = 255 - img

    # 5. Afficher img et imgResultat
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Image Originale (Gris)")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(imgResultat, cmap='gray')
    plt.title("Image Inverse (Gris)")
    plt.axis('off')

    plt.show()

```

```python
# chaging type of image to uint32 and displaying it
img_uint32 = img.astype(np.uint32)
print(f"Type de l'image originale: {img.dtype}, Type de l'image convertie: {img_uint32.dtype}")
plt.imshow(img_uint32, cmap='gray')
```

```python
# displaying image in float64
img_float64 = img.astype(np.float64) / 255.0
print(f"Type de l'image convertie en float64: {img_float64.dtype}")
plt.imshow(img_float64, cmap='gray')
plt.show()
```

```python
# displaying the pixels values
print("Valeurs des pixels de l'image originale (uint8):")
print(img)
print("Valeurs des pixels de l'image convertie (uint32):")
print(img_uint32)
```

```python
# displaying the pixels values
print("Valeurs des pixels de l'image originale (uint8):")
print(img)
print("Valeurs des pixels de l'image convertie (float64):")
print(img_float64)
```

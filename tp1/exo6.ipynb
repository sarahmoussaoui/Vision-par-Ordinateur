```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from random import randrange
```

```python
img = cv2.imread('images/fille.jpg', cv2.IMREAD_GRAYSCALE)
img[:]= img[:]/2 
h, w = img.shape
min = 255
max = 0
for y in range(h):
    for x in range(w):
        if img[y, x] > max:
            max = img[y, x]
        if img[y, x] < min:
            min = img[y, x]

for y in range(h):
    for x in range(w):
        img_brighter = [y, x] = ((img[y, x] - min) / (max - min)) * 255.


cv2.imshow('image source', img)
cv2.imshow('image bright', img_brighter)
cv2.waitKey(0)
cv2.destroyAllWindows()

hist1 = np.zeros((256, 1), np.uint64)
for y in range(h):
    for x in range(w):
        hist1[img[y, x], 0] += 1

hist2 = cv2.calcHist([img_brighter], [0], None, [256], [0, 255])

plt.figure()
plt.title('Image normalisee')
plt.xlabel('GrayScale')
plt.ylabel('nbPixels')
plt.plot(hist2)
plt.plot(hist1)
plt.xlim([0, 255])
plt.show()
```

```python
# applying filter to new image

# 1. FILTRE DE MOYENNE (Mean Filter)
# Noyau 3x3
kernel_3x3 = np.ones((3, 3), np.float32) / 9.0
mean_filtered_3x3 = cv2.filter2D(img_brighter, -1, kernel_3x3)

# 2. FILTRE MÉDIAN (Median Filter)
median_filtered_3x3 = cv2.medianBlur(img_brighter, 3)

# Afficher les résultats
cv2.imshow('Moyenne 3x3', mean_filtered_3x3)
cv2.imshow('Median 3x3', median_filtered_3x3)

hist1 = np.zeros((256, 1), np.uint64)
for y in range(h):
    for x in range(w):
        hist1[img[y, x], 0] += 1

hist2 = cv2.calcHist([mean_filtered_3x3], [0], None, [256], [0, 255])
hist2 = cv2.calcHist([mean_filtered_3x3], [0], None, [256], [0, 255])

```

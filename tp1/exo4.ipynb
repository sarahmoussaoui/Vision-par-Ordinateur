L’histogramme normalisé : aussi appelé étirement d’histogramme ça consiste à normaliser le
niveau de gris entre 0 et 255, ceci est intéressant notamment pour des images sombres mais
l’inconvénient est le bruit qui entachera d’avantage le résultat.
➢ Créez un programme qui permet de normaliser une image en utilisant la fonction suivante :


```python
'''Histogramme des niveaux de gris '''
'''pour les images qui est sauvgarder sur 16 bits uint16... pour chque c'est l image est black dans
 on pred w * h si est elle toute noir et si h= 400 et w = 400 
donc on a 160000 pixel noir et pour les histogramme il ya 8 bits !
'''

import cv2
import numpy as np
import matplotlib.pyplot as plt

# from numpy.lib.shape_base import vsplit

img = cv2.imread('images/fille.jpg', cv2.IMREAD_GRAYSCALE)
img[:]=img[:]/2
imgNorm = np.zeros((img.shape), np.uint8)
h, w = img.shape
min = 255
max = 0

for y in range(h):
    for x in range(w):
        if img[y, x] > max:
            max = img[y, x]
        if img[y, x] < min:
            min = img[y, x]

for y in range(h):
    for x in range(w):
        imgNorm[y, x] = ((img[y, x] - min) / (max - min)) * 255.

cv2.imshow('image source', img)
cv2.imshow('image normal', imgNorm)
cv2.waitKey(0)
cv2.destroyAllWindows()

hist1 = cv2.calcHist([img], [0], None, [256], [0, 255])

hist2 = cv2.calcHist([imgNorm], [0], None, [256], [0, 255])

plt.figure()
plt.title('Image normalisee')
plt.xlabel('GrayScale')
plt.ylabel('nbPixels')
plt.plot(hist2)
plt.plot(hist1)
plt.xlim([0, 255])
plt.show()
```
